### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda4:,.2f}")
```

```python
import matplotlib.pyplot as plt
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
def mostrar_grafico_barras(nombres_tiendas, ingresos):
    plt.figure(figsize=(10, 6))  # Tamaño del gráfico
    barras = plt.bar(nombres_tiendas, ingresos, color='skyblue')
    plt.title("Comparación de ingresos por tienda", fontsize=16)
    plt.xlabel("Tiendas")
    plt.ylabel("Ingresos")

    # Agregar los valores encima de cada barra
    for barra in barras:
        y = barra.get_height()
        plt.text(barra.get_x() + barra.get_width()/2, y, f'{int(y):,}', ha='center', va='bottom', fontsize=10)

    plt.tight_layout()
    plt.show()
mostrar_grafico_ingresos(nombres_tiendas, ingresos)
```

# 2. Ventas por categoría

```python
cat1 = tienda["Categoría del Producto"].value_counts()
cat2 = tienda2["Categoría del Producto"].value_counts()
cat3 = tienda3["Categoría del Producto"].value_counts()
cat4 = tienda4["Categoría del Producto"].value_counts()
print(f"Ventas por categoría Tienda 1: {cat1}")
print(f"Ventas por categoría Tienda 2: {cat2}")
print(f"Ventas por categoría Tienda 3: {cat3}")
print(f"Ventas por categoría Tienda 4: {cat4}")
```

```python
import matplotlib.pyplot as plt

def mostrar_grafico_pastel(categorias, nombre_tienda):
    plt.figure(figsize=(6, 6))
    plt.title(f'Productos vendidos por categoría - {nombre_tienda}', fontsize=14)
    plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%', startangle=90)
    plt.tight_layout()
    plt.show()
    print("\n")

mostrar_grafico_pastel(cat1, "Tienda 1")

mostrar_grafico_pastel(cat2, "Tienda 2")
mostrar_grafico_pastel(cat3, "Tienda 3")
mostrar_grafico_pastel(cat4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
promedio1 = tienda["Calificación"].mean()
promedio2 = tienda2["Calificación"].mean()
promedio3 = tienda3["Calificación"].mean()
promedio4 = tienda4["Calificación"].mean()
print(f"Calificación promedio de la Tienda 1: {promedio1}")
print(f"Calificación promedio de la Tienda 2: {promedio2}")
print(f"Calificación promedio de la Tienda 3: {promedio3}")
print(f"Calificación promedio de la Tienda 4: {promedio4}")
promedios = [promedio1, promedio2, promedio3, promedio4]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
```

```python
def mostrar_promedios_calificacion_dispersion(promedios, nombres_tiendas):
    plt.figure(figsize=(8, 5))
    plt.scatter(nombres_tiendas, promedios, color='blue', s=100, alpha=0.7)
    plt.title("Calificación promedio de los clientes por tienda", fontsize=14)
    plt.xlabel("Tiendas")
    plt.ylabel("Calificación promedio (1 a 5)")
    plt.ylim(0, 5)

    for i, promedio in enumerate(promedios):
        plt.text(i, promedio + 0.1, f'{promedio:.2f}', ha='center', va='bottom')

    plt.grid(True)
    plt.tight_layout()
    plt.show()

mostrar_promedios_calificacion_dispersion(promedios, nombres_tiendas)
```

# 4. Productos más y menos vendidos

```python
def productos_mas_menos_vendidos(df, nombre_tienda, top_n=5):
    conteo = df['Producto'].value_counts()


    top_productos = conteo.head(top_n)
    print(f"Productos más vendidos de la {nombre_tienda}: {top_productos}")
    menos_productos = conteo.tail(top_n)
    print(f"Productos menos vendidos de la {nombre_tienda}: {top_productos}")


    fig, axs = plt.subplots(1, 2, figsize=(14, 6))
    fig.suptitle(f"Productos más y menos vendidos - {nombre_tienda}", fontsize=16)


    axs[0].barh(top_productos.index[::-1], top_productos.values[::-1], color='green')
    axs[0].set_title("Más vendidos")
    axs[0].set_xlabel("Cantidad vendida")


    axs[1].barh(menos_productos.index, menos_productos.values, color='red')
    axs[1].set_title("Menos vendidos")
    axs[1].set_xlabel("Cantidad vendida")

    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()
    print("\n")

```

```python
productos_mas_menos_vendidos(tienda, "Tienda 1")
productos_mas_menos_vendidos(tienda2, "Tienda 2")
productos_mas_menos_vendidos(tienda3, "Tienda 3")
productos_mas_menos_vendidos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
prom_envio1 = tienda["Costo de envío"].mean()
prom_envio2 = tienda2["Costo de envío"].mean()
prom_envio3 = tienda3["Costo de envío"].mean()
prom_envio4 = tienda4["Costo de envío"].mean()

print(f"Envío promedio por tienda 1: {prom_envio1}")
print(f"Envío promedio por tienda 2: {prom_envio2}")
print(f"Envío promedio por tienda 3: {prom_envio3}")
print(f"Envío promedio por tienda 4: {prom_envio4}")

```

```python
promedios_envio = [prom_envio1, prom_envio2, prom_envio3, prom_envio4]
def mostrar_promedios_envio(promedios, nombres_tiendas):
    plt.figure(figsize=(8, 5))
    barras = plt.bar(nombres_tiendas, promedios, color='skyblue')
    plt.title("Costo de envío promedio por tienda", fontsize=14)
    plt.xlabel("Tiendas")
    plt.ylabel("Costo de envío promedio")
    for barra in barras:
        y = barra.get_height()
        plt.text(barra.get_x() + barra.get_width() / 2, y + 200, f'{y:,.2f}', ha='center', va='bottom')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
    print("\n")

mostrar_promedios_envio(promedios_envio, nombres_tiendas)
```

# Informe Final de Análisis de Tiendas

##Introducción:

El análisis tiene como propósito ayudar al Sr. Juan a decidir en cuál tienda debería vender sus productos. Para esto, se han considerado diferentes aspectos claves del rendimiento de cada tienda, como los ingresos generados, la popularidad de las categorías, la satisfacción de los clientes, los productos más vendidos y el costo promedio del envío. Este estudio comparativo abarca cuatro tiendas distintas, y se basa en datos reales obtenidos de sus registros de ventas.

##Análisis y Resultados

> Ingresos Totales: La Tienda 1 es la que más ingresos ha generado, aunque las demás no se quedan muy atrás.

> Categorías Más Vendidas: Todas las tiendas comparten una estructura de ventas similar:
*   Muebles y Electrónicos son las categorías más populares.
*   Libros, Instrumentos musicales y Artículos para el hogar son las menos vendidas.

Esto indica que, independientemente de la tienda, los productos del hogar tienen alta demanda.

> Calificaciones Promedio de Clientes: Aquí se observa que la Tienda 3 tiene a los clientes más satisfechos, lo que sugiere buena atención o productos de calidad.

> Productos Más Vendidos
Aunque los productos específicos más vendidos varían entre tiendas, la mayoría están relacionados a muebles y electrodomésticos.

> Costo Promedio de Envío: La Tienda 4 tiene el envío más económico, lo que podría atraer a más compradores sensibles al precio final.

##Conclusión y Recomendación
Después de analizar los datos, se recomienda que el Sr. Juan venda sus productos en la Tienda 3.
Esta tienda ofrece el mejor equilibrio entre factores clave:


*   Alta satisfacción del cliente (mejor calificación promedio).

*   Buen volumen de ventas en categorías populares.

*   Ingresos considerables, muy cerca de los de la Tienda 1.

*   Costos de envío razonables (más bajos que en la Tienda 1 y 2).

Por otro lado, la tienda que menos conviene mantener es la Tienda 1. A pesar de que genera mayores ingresos, tiene:

*   La peor calificación de clientes, lo que puede afectar la reputación a largo plazo.

*   El costo de envío más alto, lo que puede desincentivar algunas compras.

*   Una estructura de ventas muy similar a las otras tiendas, por lo que sus ventajas no son exclusivas.
